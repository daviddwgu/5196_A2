### 1.3 check the date is correct or not

```python
# make a function to check the date format is correct or not
def isVaildDate(date):
    try:
        time.strptime(date, "%Y-%m-%d")
        return True
    except:
        return False
```

```python
dirty_data["valid_date"] = dirty_data["Departure Date"].apply(isVaildDate)
dirty_data[dirty_data["valid_date"] == False]
```

- Look Though the problem data, we can found there are two type of error. The first type of problem is date format is not correct like 2018-20-05 , the correct format should be year-month-day. The second type of problem is date is not exist like 2018-02-30.

```python
# make a function to shift the first type of problem format into correct format. 
def shift_date(date):
    year = date.split("-")[0].strip()
    month = date.split("-")[2].strip()
    day = date.split("-")[1].strip()
    return year + "-" + month + "-" + day
```

```python
# extract the month of date data, if the format data is not correct, the month will be more than 12
coltitle = dirty_data['Departure Date'].apply(lambda s: pd.Series({'month': int(s.split('-')[1].strip())}))
# Add the columns to the titanic dataframe
dirty_data = pd.concat([dirty_data, coltitle], axis=1)
```

```python
# check the problem format data
dirty_data[(dirty_data["month"] > 12)]
```

```python
# Apply shift_date function to those problem format data
dirty_data.loc[(dirty_data["month"] > 12), "Departure Date"] = dirty_data["Departure Date"].apply(shift_date)
```

```python
# Check the date data with function valid_date to see the first type of problem is gone or not.
dirty_data["valid_date"] = dirty_data["Departure Date"].apply(isVaildDate)
dirty_data[dirty_data["valid_date"] == False]
```

- look though the problem data, there is only type 2 problem left. For 2018-04-31, 2018-04-30 and 2018-05-01 are both weekday, so changing the 2018-04-31 to either side of date is not affecting the fare. For 2018-02-30, 2018-02-28 and 2018-03-01 are both weekday, so changing 2018-02-30 to either side of date also is not affecting the fare. For 2018-06-31, 2018-06-30 and 2018-07-01 are both weekend, so changing 2018-06-30 to eigher side of date also is not affecting the fare.

```python
# make change to those problem date
dirty_data["Departure Date"].replace({'2018-02-30':'2018-02-28', '2018-06-31':'2018-06-30', '2018-04-31':'2018-04-30'},inplace=True)
```

### 1.4 Check the distance have problem or not

```python
# define function for identify the node id base on the latitude and longitude
def found_node_id(latitude, longitude):
    result = nodes[(round(nodes["lat"],7)==latitude)&(round(nodes["lon"],7)==longitude)].values[0][0]
    return result
```

```python
# identify the origin node id and destination node id
origin_id_list = []
for i in range(len(dirty_data)):
    origin_id_list.append(found_node_id(round(dirty_data.loc[i]["Origin Latitude"],7), round(dirty_data.loc[i]["Origin Longitude"],7)))
destin_id_list = []
for i in range(len(dirty_data)):
    destin_id_list.append(found_node_id(round(dirty_data.loc[i]["Destination Latitude"],7), round(dirty_data.loc[i]["Destination Longitude"],7)))
    
dirty_data["origin_id"] = origin_id_list
dirty_data["destin_id"] = destin_id_list
```

```python
# Check is there null value exist or not 
dirty_data.isnull().sum()
```

- Since there is no null value exist, it mean the changing latitude data without problem, it still can identify the nodes id out. 

```python
def traveltime(path,edges):
    time = 0 
    for i in range(len(path)-1):
        distance = edges[((edges.u==path[i])&(edges.v==path[i+1]))|((edges.v==path[i])&(edges.u==path[i+1]))]["distance(m)"].values[0]
        speed = edges[((edges.u==path[i])&(edges.v==path[i+1]))|((edges.v==path[i])&(edges.u==path[i+1]))]["speed(km/h)"].values[0]
        path_time = (distance*1.0*3600)/(speed*1000)
        time += path_time
    return time                  
```

```python
# insert edges.csv data into program for calculate the distance base on shortest distance by Dijkstra algorithm.
graph = Graph()
for i in range(len(edges)):
    graph.add_edge(edges.iloc[i, 1], edges.iloc[i,2], {'distance': edges.iloc[i,3]})
    graph.add_edge(edges.iloc[i, 2], edges.iloc[i,1], {'distance': edges.iloc[i,3]})
cost_func = lambda u, v, e, prev_e: e['distance']
```

```python
# calculate the distance of each rows by Dijkstra algorithm, and creat a column for calculated distance
distance_data = []
time_data = []
for i in range(len(dirty_data)):
    result = find_path(graph, dirty_data.iloc[i,]["origin_id"], dirty_data.iloc[i,]["destin_id"] , cost_func=cost_func)
    distance_data.append(result[3])
    time_data.append(traveltime(result[0],edges))
print(len(distance_data))
print(len(time_data))
print(distance_data)
print(time_data)
dirty_data["calculate_distance"] = distance_data
dirty_data["calculate_time"] = time_data
```

```python
# check which row's Journey Distance data is not match with calculated distant 
dirty_data[dirty_data["Journey Distance(m)"] != dirty_data["calculate_distance"]]
```

```python
# replace those problem row distance with calculate distance
dirty_data.loc[(dirty_data["Journey Distance(m)"] != dirty_data["calculate_distance"]), "Journey Distance(m)"] = dirty_data["calculate_distance"]
```

```python
# Final check the row's Journey Distance data is match with calculated distant or not 
dirty_data[dirty_data["Journey Distance(m)"] != dirty_data["calculate_distance"]]
```

```python
del dirty_data["origin_id"]
del dirty_data["destin_id"]
del dirty_data["calculate_distance"]
```

### 2.2 Handling the missing value in Fare
- Since we are using linear regression to predict the null value in fare, the rate to calcuate fare for 3 uber type will be different, so we will make 3 models, and fare calculation will base on traveling time, distance, and date check ( weekday or weekend), and time check(in 3 time period).
- Since we only have around 100 row for linear regression, there is not enough data to make arrcurate prediction, so we adding part 1 data(cleaned dirty_data) into dataset for calculation

```python
# plan to combine the missing value data with cleaned dirty data, so make copy of missing_data
missing_data_copy = missing_data.copy()
```

```python
# load the clean dirty data, and combine with missing value data
dirty_data = pd.read_csv("29910226_dirty_data_solution.csv")
dirty_data.rename(columns={'Unnamed: 0':'id'}, inplace = True)
dirty_data_copy = dirty_data.copy()
model_data = pd.concat([missing_data_copy,dirty_data_copy], axis=0, ignore_index=True, sort=False)
model_data
```

```python
# combination causes all of id of cleaned dirty data have been move to the no name column, so move the id back, and delate the no name column
model_data.rename(columns={' ':'id_2'}, inplace = True)
model_data.loc[(model_data["id"].isnull()),"id"] = model_data['id_2']
del model_data['id_2']
```

```python
# check the dataset
model_data
```

```python
# check existence of duplicated row
model_data[model_data['id'].duplicated()]
```

```python
# define a function to check the date is weekday or not, and apply function to both missing data and copy of missing data
def check_weekday(date):
    weekday = pd.to_datetime(date).weekday()
    if weekday < 5:
        return 0
    else:
        return 1
    
missing_data["check_weekday"] = missing_data["Departure Date"].apply(check_weekday)
model_data["check_weekday"] = model_data["Departure Date"].apply(check_weekday)
```

```python
# define a function to check the time is belong to which section, and apply function to both missing data and copy of missing data.
def check_time(time):
    time = datetime.datetime.strptime(time, '%H:%M:%S')
    if (time >= datetime.datetime.strptime("06:00:00", '%H:%M:%S')) & (time <= datetime.datetime.strptime("11:59:59", '%H:%M:%S')):
        return 0
    elif (time >= datetime.datetime.strptime("12:00:00", '%H:%M:%S')) & (time <= datetime.datetime.strptime("20:59:59", '%H:%M:%S')):
        return 1
    else:
        return 2
missing_data["check_time"] = missing_data["Departure Time"].apply(check_time)
model_data["check_time"] = model_data["Departure Time"].apply(check_time)
```

```python
# output the copy for outliers model training
model_data.dropna(subset=['Fare$'],axis=0,inplace=True)
model_data.to_csv('model_data.csv',index=False)
```

#### Linear Regression

```python
# Seperating the model_data and null_value row in missing_data by uber type

# Model data for different uber type
model_data_0 = model_data[model_data["Uber Type"]==0].copy()
model_data_1 = model_data[model_data["Uber Type"]==1].copy()
model_data_2 = model_data[model_data["Uber Type"]==2].copy()

# Seperate data into training data and test data
x_train_0, x_test_0, y_train_0, y_test_0 = train_test_split(model_data_0.loc[:,("Journey Distance(m)", "Travel Time(s)", "check_weekday", "check_time")], model_data_0['Fare$'], random_state = 1)
x_train_1, x_test_1, y_train_1, y_test_1 = train_test_split(model_data_1.loc[:,("Journey Distance(m)", "Travel Time(s)", "check_weekday", "check_time")], model_data_1['Fare$'], random_state = 1)
x_train_2, x_test_2, y_train_2, y_test_2 = train_test_split(model_data_2.loc[:,("Journey Distance(m)", "Travel Time(s)", "check_weekday", "check_time")], model_data_2['Fare$'], random_state = 1)

# extract the null value row base on uber type
null_value_0 = missing_data[(missing_data['Fare$'].isnull())&(missing_data["Uber Type"]==0)].copy()
null_value_1 = missing_data[(missing_data['Fare$'].isnull())&(missing_data["Uber Type"]==1)].copy()
null_value_2 = missing_data[(missing_data['Fare$'].isnull())&(missing_data["Uber Type"]==2)].copy()
```

```python
# Fitting the train data to model
lm_for_impute_0 = LinearRegression()
lm_for_impute_0.fit(x_train_0, y_train_0) 

lm_for_impute_1 = LinearRegression()
lm_for_impute_1.fit(x_train_1, y_train_1) 

lm_for_impute_2 = LinearRegression()
lm_for_impute_2.fit(x_train_2, y_train_2) 
```

```python
# Show the R-Spuare value for each Uber type model to check the model fitting well with test group or not
print("R-Spuare for Uber type 0 model is", lm_for_impute_0.score(x_test_0, y_test_0))
print("R-Spuare for Uber type 1 model is", lm_for_impute_1.score(x_test_1, y_test_1))
print("R-Spuare for Uber type 2 model is", lm_for_impute_2.score(x_test_2, y_test_2))
```

```python
# Make prediction to null value of each uber type
print("null value prediction for uber type 0",lm_for_impute_0.predict(null_value_0.loc[:,("Journey Distance(m)", "Travel Time(s)", "check_weekday", "check_time")]))
print("null value prediction for uber type 1",lm_for_impute_1.predict(null_value_1.loc[:,("Journey Distance(m)", "Travel Time(s)", "check_weekday", "check_time")]))
print("null value prediction for uber type 2",lm_for_impute_2.predict(null_value_2.loc[:,("Journey Distance(m)", "Travel Time(s)", "check_weekday", "check_time")]))
```

```python
# Since there is no negative value, impute the value to dataset
null_value_0.loc[:,'Fare$'] = lm_for_impute_0.predict(null_value_0.loc[:,("Journey Distance(m)", "Travel Time(s)", "check_weekday", "check_time")])
null_value_1.loc[:,'Fare$'] = lm_for_impute_1.predict(null_value_1.loc[:,("Journey Distance(m)", "Travel Time(s)", "check_weekday", "check_time")])
null_value_2.loc[:,'Fare$'] = lm_for_impute_2.predict(null_value_2.loc[:,("Journey Distance(m)", "Travel Time(s)", "check_weekday", "check_time")])
```

```python
# impute the predict value back to the missing data

# Uber type 0
for i in range(len(null_value_0)):
    temp_id = null_value_0.iloc[i,0]
    missing_data.loc[(missing_data["id"]==temp_id), 'Fare$'] = null_value_0.iloc[i,13]
    
# Uber type 1
for i in range(len(null_value_1)):
    temp_id = null_value_1.iloc[i,0]
    missing_data.loc[(missing_data["id"]==temp_id), 'Fare$'] = null_value_1.iloc[i,13]
    
# Uber type 2
for i in range(len(null_value_2)):
    temp_id = null_value_2.iloc[i,0]
    missing_data.loc[(missing_data["id"]==temp_id), 'Fare$'] = null_value_2.iloc[i,13]
```

```python
# Check is there null value exist or not 
missing_data.isnull().sum()
```

```python
del missing_data["check_weekday"]
del missing_data["check_time"]
```
